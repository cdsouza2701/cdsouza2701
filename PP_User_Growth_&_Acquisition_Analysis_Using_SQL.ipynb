#### 4.3.1.1 Growth Questions- User Table

1️⃣ Growth Questions (MOST IMPORTANT)

**Business question:**

Are we growing or slowing down?

**What you can learn:**

- User growth over time

- Seasonality

- Sudden drops or spikes

```python
pd.read_sql("""
SELECT
  strftime('%Y-%m', signup_date) AS month,
  COUNT(*) AS new_users
FROM users
GROUP BY 1
ORDER BY 1;
 """, conn)
```

```python
# See raw dates
pd.read_sql("""
SELECT signup_date
FROM users""", conn)
```

```python
# See year only
pd.read_sql("""
SELECT signup_date, strftime('%Y', signup_date) AS Year
FROM users
""", conn)
```
