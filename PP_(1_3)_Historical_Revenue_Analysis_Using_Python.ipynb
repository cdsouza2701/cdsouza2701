#Part 2: Working on solution

```python
#Load data
transactions = (pd.read_csv('transactions.csv', parse_dates=["order_date"])). sort_values(by="order_date")

price_history = (pd.read_csv('price_history.csv', parse_dates=['effective_date'])).sort_values(by='effective_date')
```
