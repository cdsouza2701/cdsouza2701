#1️⃣ Importing libraries

```python
from bs4 import BeautifulSoup
import requests
```

**requests:**
Handles HTTP communication (sending requests to websites and receiving responses).

**BeautifulSoup** (from bs4):
Parses raw HTML into a Python object that you can search, navigate, and extract data from.

#2️⃣ Defining the target URL

```python
url = 'https://quotes.toscrape.com'
```

#3️⃣ Sending an HTTP GET request

```python
page_to_scrape = requests.get(url)
```

- Script sends a GET request to the server

- The server responds with:

  - HTML content

  - Status code (e.g. 200, 404, 500): 200 → success, Anything else → request failed

The page_to_scrape object contains everything the server sends back.

```python
print(page_to_scrape.status_code)
```

#4️⃣ Parsing the HTML content

```python
soup = BeautifulSoup(page_to_scrape.text, 'html.parser')
```

- **response.text:** raw HTML as a long string

- **"html.parser":** tells Beautiful Soup how to read the HTML

- **soup:** a parsed representation of the page (DOM tree)

```python
print(soup.prettify)
```

#5️⃣ Finding multiple elements on the page

```python
quotes = soup.find_all("span", class_="text")
authors = soup.find_all("small", class_="author")
```

What find_all() does

   - Searches the HTML tree

   - Returns a list of matching elements

Why these selectors work:

If you inspect the website:

Quotes are inside: `<span class="text">`

Authors are inside: `<small class="author">`

So Beautiful Soup pulls all matching elements into lists.

# 6️⃣ Looping through quotes and authors together

```python
for quote, author in zip(quotes, authors):
    print(quote.text)
    print(author.text)
```

**Why zip()?**

- quotes and authors are separate lists

- zip() pairs:

  - first quote with first author

  - second quote with second author

  - etc.

This prevents mismatches.

**.text explained**

`quote.text`

- Extracts only the text

- Removes HTML tags

- Returns a clean Python string
